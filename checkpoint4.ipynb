## Desafio 1 — Seleção de Projetos de Mitigação

Selecionar projetos que maximizem o benefício ambiental sem ultrapassar o orçamento, usando três paradigmas de algoritmos.

### Considerações Iniciais
- Os projetos são representados por pares (custo, benefício).
- O orçamento é um inteiro.
- Os algoritmos implementados são: Força Bruta, Meet-in-the-Middle, Programação Dinâmica.
- Decorators são usados para medir tempo de execução.

```python
# Imports e geração de dados
import numpy as np
from mitigation_selection import brute_force, meet_in_the_middle, dp_bottom_up
np.random.seed(42)

def gerar_projetos(n):
    custos = np.random.randint(2, 21, n)
    beneficios = np.random.randint(10, 121, n)
    return list(zip(custos, beneficios))
```

```python
# Teste dos algoritmos para diferentes valores de n
import matplotlib.pyplot as plt
import time

ns = range(10, 26)
budget = 50
resultados = {'Brute Force': [], 'Meet-in-the-Middle': [], 'DP': []}
tempos = {'Brute Force': [], 'Meet-in-the-Middle': [], 'DP': []}

for n in ns:
    projetos = gerar_projetos(n)
    start = time.time()
    brute_force(projetos, budget)
    tempos['Brute Force'].append((time.time()-start)*1000)
    start = time.time()
    meet_in_the_middle(projetos, budget)
    tempos['Meet-in-the-Middle'].append((time.time()-start)*1000)
    start = time.time()
    dp_bottom_up(projetos, budget)
    tempos['DP'].append((time.time()-start)*1000)

plt.figure(figsize=(8,5))
for k in tempos:
    plt.plot(ns, tempos[k], label=k)
plt.xlabel('Número de projetos (n)')
plt.ylabel('Tempo de execução (ms)')
plt.title('Desempenho dos Algoritmos – Seleção de Projetos')
plt.legend()
plt.show()
```

### Explicação dos Algoritmos
- **Força Bruta**: Testa todos os subconjuntos possíveis, garantindo a solução ótima, porém com alto custo computacional.
- **Meet-in-the-Middle**: Divide o problema em duas partes, combinando soluções parciais para maior eficiência.
- **Programação Dinâmica**: Utiliza tabela para armazenar soluções intermediárias, reduzindo drasticamente o tempo de execução.

## Desafio 2 — Janela de Qualidade do Ar

Encontrar o período contínuo de dias com maior soma de pontuações ambientais, usando três paradigmas de algoritmos.

### Considerações Iniciais
- Pontuações diárias variam de -10 a 10.
- Algoritmos: Força Bruta, Divisão e Conquista, Kadane.
- Decorators para medir tempo de execução.

```python
# Imports e geração de dados
from air_quality_window import brute_force as aq_brute_force, divide_and_conquer, kadane

def gerar_pontuacoes(n):
    return np.random.randint(-10, 11, n).tolist()
```

```python
# Teste dos algoritmos para diferentes valores de n
ns = [200, 400, 800, 1200, 1600]
tempos_aq = {'Brute Force': [], 'Divide and Conquer': [], 'Kadane': []}

for n in ns:
    pontuacoes = gerar_pontuacoes(n)
    start = time.time()
    aq_brute_force(pontuacoes)
    tempos_aq['Brute Force'].append((time.time()-start)*1000)
    start = time.time()
    divide_and_conquer(pontuacoes)
    tempos_aq['Divide and Conquer'].append((time.time()-start)*1000)
    start = time.time()
    kadane(pontuacoes)
    tempos_aq['Kadane'].append((time.time()-start)*1000)

plt.figure(figsize=(8,5))
for k in tempos_aq:
    plt.plot(ns, tempos_aq[k], label=k)
plt.xlabel('Número de dias (n)')
plt.ylabel('Tempo de execução (ms)')
plt.title('Desempenho dos Algoritmos – Janela de Qualidade do Ar')
plt.legend()
plt.show()
```

### Explicação dos Algoritmos
- **Força Bruta**: Avalia todas as janelas possíveis, garantindo a solução ótima, mas com alto custo computacional.
- **Divisão e Conquista**: Divide o problema em partes menores, combinando soluções para maior eficiência.
- **Kadane**: Algoritmo eficiente que encontra a soma máxima em tempo linear.

## Conclusões

Os resultados mostram que algoritmos de programação dinâmica e divisão e conquista são muito mais eficientes e escaláveis que força bruta, especialmente para grandes valores de n. A escolha do paradigma impacta diretamente a viabilidade computacional das soluções em contextos reais como a COP30.
